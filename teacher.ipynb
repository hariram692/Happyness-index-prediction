## Pre-processing

```python
from sklearn.preprocessing import LabelEncoder 

lab=LabelEncoder()

data2['1. Is salary being given on time?']=lab.fit_transform(data2['1. Is salary being given on time?'])
data2['2. What about the teaching and learning process?']=lab.fit_transform(data2['2. What about the teaching and learning process?'])
data2['3. How is the infrastructure?']=lab.fit_transform(data2['3. How is the infrastructure?'])
data2['4.  Teaching experience']=lab.fit_transform(data2['4.  Teaching experience'])
data2['5. How long have you been working in the same school?']=lab.fit_transform(data2['5. How long have you been working in the same school?'])
data2['6. Is teaching done on activity based?']=lab.fit_transform(data2['6. Is teaching done on activity based?'])
data2['7.How well can students speak and understand English?']=lab.fit_transform(data2['7.How well can students speak and understand English?'])
```

## Data Splitting

```python
x=data2.iloc[:,data2.columns!='Happiness Index']
y=data2.iloc[:,data2.columns=='Happiness Index']

#x.shape

#x.head()

x.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,recall_score, roc_curve, auc
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)

xtrain.head()

ytrain.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor,RandomForestRegressor
from sklearn.ensemble import StackingRegressor
#base models
base_models=[
    ('ABR',AdaBoostRegressor(n_estimators=50)),
    ('GBR',GradientBoostingRegressor(n_estimators=50)),
    ('RFR',RandomForestRegressor(n_estimators=50))
]
#Final Model
final_model=LinearRegression()

#stacking Regressor
classifier=StackingRegressor(estimators=base_models,final_estimator=final_model)
classifier.fit(xtrain,ytrain)
ypred=classifier.predict(xtest)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
#Evaluation of Model
print("R2 Score " ,r2_score(ytest,ypred))
mse=mean_squared_error(ytest,ypred)
mae=mean_absolute_error(ytest,ypred)
print("MSE:",mse)
print("MAE:",mae)
print("RMSE :",np.sqrt(mse))
```

```python
import pickle
filename = 'teacher.sav'
pickle.dump(classifier, open(filename, 'wb'))
# loading the saved model
loaded_model = pickle.load(open('teacher.sav', 'rb'))
```
